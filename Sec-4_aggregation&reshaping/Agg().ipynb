# Agg() method

```python
import pandas as pd
```

```python
stores=pd.read_csv(r"E:\Devu\USB Drive\coding\pandas-from-scratch\notebooks\Sec-4_aggregation&reshaping\stores.csv")
stores.head()
```

```python
stores.groupby(["city","state"]).agg("sum")
```

```python
stores.groupby(["city","state"]).agg("sum","count")
```

```python
stores.groupby(["city","state"]).agg({"cluster":["sum","mean"]})
```

```python
stores.groupby(["city","state"]).agg(cluster_sum=("cluster","sum"),
                                     cluster_mean=("cluster","mean"))
```
