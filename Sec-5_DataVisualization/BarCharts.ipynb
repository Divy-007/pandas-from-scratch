# Bar charts 

```python
import pandas as pd 
```

```python
store = pd.read_csv(r"F:\coding\pandas-from-scratch\notebooks\Sec-5_DataVisualization\stores.csv")
store.head()

```

```python
store.loc[store["cluster"]].plot(
    kind="bar",
    rot=45
)
```

```python
store.query("store_nbr in [1,2,3,4] ").plot.barh(rot=45,
        figsize=(6,6))
```

# grouped bar charts

```python
store.groupby(["city","state"]).agg({"store_nbr":["sum"],
          "cluster":["mean"]}).apply(lambda x: x*100/sum(x),axis=1).plot.bar(stacked=True,legend=False,figsize=(10,10))
```
