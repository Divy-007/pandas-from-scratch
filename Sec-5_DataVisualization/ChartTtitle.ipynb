```python
import pandas as pd
import matplotlib as plt
# plt.style.use("fivethirtyeight")
```

```python
oil = pd.read_csv(r"F:\coding\pandas-from-scratch\notebooks\Sec-5_DataVisualization\oil.csv")
oil

```

```python
oil.astype({"date":"datetime64"}).set_index("date").loc["2014"].plot(
    title="oil per barrel price",
    xlabel="year",
    ylabel="oil_price",
    # color="#4e351f",
    cmap="Pastel2",
    style=".-",
    grid=True,
    figsize=(6,6)
).legend(bbox_to_anchor=(1.2,1.2));
```
