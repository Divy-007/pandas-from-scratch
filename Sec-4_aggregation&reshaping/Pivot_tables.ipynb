# pivot_tables() method

```python
import pandas as pd
```

```python
stores=pd.read_csv(r"E:\Devu\USB Drive\coding\pandas-from-scratch\notebooks\Sec-4_aggregation&reshaping\stores.csv")
stores.head()
```

```python
stores.query('city in ["Quito", "Santo Domingo"] and store_nbr>20 ').pivot_table(index="store_nbr",
                  columns="city",
                  values="cluster",
                  aggfunc="sum",
                  margins=True)
```

```python
stores.query('city in ["Quito", "Santo Domingo"] and store_nbr>20 ').pivot_table(index="store_nbr",
                  columns="city",
                  values="cluster",
                  aggfunc=(["sum","mean"]),
                  margins=True)
```
